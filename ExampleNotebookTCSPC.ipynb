```python
#import
from tcspc import TCSPC
import matplotlib.pyplot as plt
import numpy as np
```

```python
#set path
path = "C:\\Users\\peo0005\\OneDrive - Auburn University\\Documents\\GitHub\\tcspc\\aAS_TEST3_RH80.dat"

#import dataset from one file
rh80data = TCSPC(path)
```

```python
#plot
#channel = channel to be plotted
#bgchannel = channel to subtract, or None if no subtraction
#xlim = x-axis time limit in ns, or none if full axis
#smooth = savitzky-golay smoothing parameter to also plot, or None if no smoothing
rh80data.plot(channel = 2, bgchannel = 3, xlim=[0,50],smooth = 100)
```

```python
#get one trace from a data object, including optional bgsubtraction and smoothing
rh80_smoothed = rh80data.getData(channel = 2, bgchannel = 3, smooth = 100)

#now you can do whatever you want with the trace
print(rh80_smoothed)
plt.plot(rh80_smoothed['time'],rh80_smoothed['counts'])

#sum of photon counts
print(np.sum(rh80_smoothed['counts']))
```
